```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
openai = OpenAI()
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
```

```python
personas = {
    "Optimist": "You are an optimist in a casual panel discussion. Keep responses to 2-3 sentences max. Speak naturally and conversationally like a real person. Focus on positive angles and solutions, but don't lecture. React directly to what others just said.",
    "Pessimist": "You are a pessimist in a casual panel discussion. Keep responses to 2-3 sentences max. Speak naturally and conversationally like a real person. Point out flaws and risks, maybe roll your eyes at the optimist's naivety, but keep it snappy and real.",
    "Realist": "You are a realist in a casual panel discussion. Keep responses to 2-3 sentences max. Speak naturally and conversationally like a real person. Find middle ground between the extremes, acknowledge valid points from both sides, and keep it grounded in facts."
}
```

```python
def get_ai_response(persona_instruction, conversation_transcript, topic):
    
    prompt_context = f"""You are in a panel discussion about: '{topic}'.

Conversation so far:
{conversation_transcript}

Now it's your turn to speak. Respond to what the others just said - agree, disagree, add your perspective, or challenge their points. Make it feel like you're actually talking TO them, not just answering the original question again. Keep it natural and conversational."""
    
    messages = [
        {"role": "system", "content": persona_instruction},
        {"role": "user", "content": prompt_context}
    ]
    
    response = openai.chat.completions.create(
        model="gpt-5-nano",
        messages=messages,   
    )
    
    return response.choices[0].message.content
```

```python
topic = "Flying Cars"

starting_question = "Welcome. Do you think flying cars will become real in future?"

transcript = f"Moderator: {starting_question}\n\n"

speaker_order = ["Optimist", "Pessimist", "Realist"]
rounds = 2

print(f"--- Starting Conversation on: {topic} ---\n")
print(f"Moderator: {starting_question}\n")

for round_num in range(rounds):
    print(f"*** Round {round_num + 1} ***")
    
    for speaker_name in speaker_order:
        
        current_persona = personas[speaker_name]
        
        reply = get_ai_response( 
            persona_instruction=current_persona, 
            conversation_transcript=transcript, 
            topic=topic
        )
        
        print(f"{speaker_name}: {reply}\n")
        
        transcript += f"{speaker_name}: {reply}\n\n"

print("--- Conversation Ended ---")
```
